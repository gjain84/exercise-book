# Cleaning the data


```python
del df["Category Name"]
del df["Item Description"]
del df["County"]
```

```python
df
```

```python
cols = ["State Bottle Cost", "State Bottle Retail", "Sale (Dollars)"]
for col in cols:
    df[col] = df[col].apply(lambda x: float(x[1:]))
```

```python
df
```

```python
df["Date"] = pandas.to_datetime(df["Date"], format="%m/%d/%Y")
```

```python
df.dropna(inplace=True)
```

```python
df.dtypes
```

```python
df["County Number"] = df["County Number"].astype(int)
df["Category"] = df["Category"].astype(int)
```

```python
df.dtypes
```

# Filter the data

```python
import datetime
```

```python
dates = df.groupby(by=["Date"], as_index=False)
dates
```

```python
dates = df.groupby(by=["Store Number"], as_index=False)
dates = dates.agg({"Date": [numpy.min, numpy.max]})
#dates.columns = [' '.join(col).strip() for col in dates.columns.values]
dates
```

```python
opened_early = dates[('Date','amin')] < datetime.datetime(2015, 1, 15)
opened_late = dates[('Date','amax')] > datetime.datetime(2015, 12, 10)
```

```python
opened_all = opened_early & opened_late
opened_all
```

```python
 dates[opened_all]     
        
dates[opened_all].max()
```

```python
true_values = dates[opened_all== True] 
```

```python
df2 = df.merge(dates[opened_all], left_on='Store Number', right_on='Store Number')
df2
```

# Margin, Price per Bottle and Price per Liter

```python
df2["Margin"] = (df2["State Bottle Retail"] - df2["State Bottle Cost"]) * df2["Bottles Sold"]
df2["Price per Liter"] = df2["Sale (Dollars)"] / df2["Volume Sold (Liters)"]
df2["Price per Bottle"] = df2["Sale (Dollars)"] / df2["Bottles Sold"]
df2.head()
```

 # Sales Quarters

```python
df.sort_values(by=["Store Number", "Date"], inplace=True)
start_date = pandas.Timestamp("20150101")
end_date = pandas.Timestamp("20151231")
mask = (df['Date'] >= start_date) & (df['Date'] <= end_date)
sales = df[mask]
```
