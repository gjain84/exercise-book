```python
for word in range(1,len(reviews.columns)+1):
    cleaner = []
    for x in eval('reviews.col_'+str(word)):
        cleaner.append(x)
    reviews['col_'+str(word)] = cleaner
```

```python
for word in range(1,len(reviews.columns)+1):
    cleaner = []
    for x in eval('reviews.col_'+str(word)):
        cleaner.append(re.sub(r"[^a-zA-Z0-9 ]", "", x))
    reviews['col_'+str(word)] = cleaner
reviews
```

# Grid Search and Regression  

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.grid_search import GridSearchCV
```

```python
X = final_df.ix[:,5:]

y = final_df.rating.values.astype(int)
```

```python
count_vec = CountVectorizer()
count_vec.fit(X)
X = pd.DataFrame(count_vec.transform(X).todense())
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

## Tree

```python
d_tree = DecisionTreeClassifier(class_weight='balanced')
d_tree.fit(X_train, y_train)
```

```python
d_tree.score(X_test, y_test)
```

```python
params = {'max_depth':[3,5,10,20], 'max_features':[1,2,3,'auto'], 
              'max_leaf_nodes':[5,6,7,8,9,10], 'min_samples_leaf':[1,3,5,7,10],
              'min_samples_split':[2,5,7]}
scoring = 'mean_squared_error'
```

```python
#Grid
g_tree = GridSearchCV(d_tree, params, scoring=scoring, verbose=True, n_jobs=-1)
```

```python
g_tree.fit(X, y)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.cross_validation import cross_val_score
predictions = d_tree.predict(X_test)
accuracy_score(y_test, predictions)
```

```python
cross_val_score(d_tree, X, y, cv=10)
```
