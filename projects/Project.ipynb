```python
wiki_page = 'https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States'
dfs = pandas.read_html(wiki_page, index_col=[0], header=0)
```

```python
dfs[0]
```

```python
populations = dfs[0]
def is_number(x):
    try:
        int(x)
        return True
    except ValueError:
        return False

correct_answers = []
for row_idx in populations.index:
    if is_number(populations.loc[row_idx, "Statehood[B]"]):
        correct_answers.append(int(populations.loc[row_idx, "Statehood[B]"]))
    else:
        correct_answers.append(int(populations.loc[row_idx, "Population (2015 est)[15]"]))

populations['CleanPopulation'] = correct_answers
populations['CleanPopulation']
```

```python
sat_scores_with_population = sat_scores.merge(populations[["Abbr.","CleanPopulation"]], left_index=True, right_on='Abbr.').copy()
sat_scores_with_population
```

```python
sat_scores_with_population['TotalPeople'] = (sat_scores_with_population.Rate/100) * sat_scores_with_population.CleanPopulation
```
